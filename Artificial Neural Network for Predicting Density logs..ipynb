```python
import tensorflow as tf
import keras
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras import Sequential
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import L2
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
from sklearn.metrics import r2_score
```

```python
log1=pd.read_csv('C:/Users/Paa Kwasi/Desktop/Python Notebooks/CONGER # E9-i.csv').dropna()
```

```python
log1
```

```python
log1=log1[(log1['RHOB']>2.35)&(log1['RHOB']<2.7)]
```

```python
log1
```

```python
X=log1[['DEPTH','DPOR','GR','NPOR']]
y=log1['RHOB']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y, random_state=1)
```

```python
model= Sequential([
    Dense(units=50, activation='relu',kernel_regularizer=L2(0.001)),
    Dense(units=20, activation='relu',kernel_regularizer=L2(0.001)),
    Dense(units=5, activation='relu',kernel_regularizer=L2(0.001)),
    Dense(units=1, activation='linear',kernel_regularizer=L2(0.001))
])
```

```python
model.compile(loss=MeanSquaredError(),optimizer=Adam(0.0001),metrics=['mean_squared_error'])
```

```python
model.fit(X_train,y_train, epochs=10000)
```

```python
model.evaluate(X_test,y_test)
```

```python
r2_score(y_test,model.predict(X_test))
```
