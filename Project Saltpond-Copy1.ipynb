```python
random_state=[x for x in range(200)]
mse_test=[]
mse_train=[]
train_score=[]
test_score=[]
for i in random_state:
    from sklearn.model_selection import train_test_split
    from sklearn.preprocessing import MinMaxScaler
    scaler=MinMaxScaler()
    X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=i)
    X_train=scaler.fit_transform(X_train)
    X_test=scaler.transform(X_test)
    from sklearn.linear_model import LinearRegression
    lr=LinearRegression()
    lr.fit(X_train,y_train)
    from sklearn.metrics import mean_squared_error
    train_score.append(lr.score(X_train,y_train))
    test_score.append(lr.score(X_test,y_test))
    mse_train.append(mean_squared_error(y_train,lr.predict(X_train)))
    mse_test.append(mean_squared_error(y_test,lr.predict(X_test)))
    df_mse_train=pd.DataFrame(mse_train,columns=['MSE_train'])
    df_mse=pd.DataFrame(mse_test, columns=["MSE_test"])
    df_train=pd.DataFrame(train_score, columns=['Training_score'])
    df_test=pd.DataFrame(test_score,columns=["Test_score"])
    df=pd.merge(df_train,df_test, how='outer', left_index=True, right_index=True)
    df=pd.merge(df,df_mse,how='outer',left_index=True, right_index=True)
    df=pd.merge(df,df_mse_train,how='outer',left_index=True, right_index=True)
```

```python
mse_test=[]
mse_train=[]
train_score=[]
test_score=[]
c= np.linspace(2,2.9,50)
for i in c:
    from sklearn.model_selection import train_test_split
    from sklearn.preprocessing import MinMaxScaler
    scaler=MinMaxScaler()
    X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=33)
    X_train=scaler.fit_transform(X_train)
    X_test=scaler.transform(X_test)
    from sklearn.svm import SVR
    svr=SVR(C=i)
    svr.fit(X_train,y_train)
    from sklearn.metrics import mean_squared_error
    train_score.append(svr.score(X_train,y_train))
    test_score.append(svr.score(X_test,y_test))
    mse_train.append(mean_squared_error(y_train,svr.predict(X_train)))
    mse_test.append(mean_squared_error(y_test,svr.predict(X_test)))
    df_mse_train=pd.DataFrame(mse_train,columns=['MSE_train'])
    df_mse=pd.DataFrame(mse_test, columns=["MSE_test"])
    df_train=pd.DataFrame(train_score, columns=['Training_score'])
    df_test=pd.DataFrame(test_score,columns=["Test_score"])
    df=pd.merge(df_train,df_test, how='outer', left_index=True, right_index=True)
    df=pd.merge(df,df_mse,how='outer',left_index=True, right_index=True)
    df=pd.merge(df,df_mse_train,how='outer',left_index=True, right_index=True)
```
